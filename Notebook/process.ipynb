```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:\Major7\AI1702\DAT301m\Chatbot\Dataset\MCQ_medical\questions_601_700.csv')
df.head()
```

```python
df["ANSWER"] = ""

for index, row in df.iterrows():
    answer = ""
    correct = row["CORRECT"]
    choices = ["A", "B", "C", "D", "E", "F"]

    for choice in choices:
        if choice in correct:
            if len(answer) > 0:
                answer += ", "
            answer += str(row[choice])

    df.at[index, "ANSWER"] = answer

df.head()
```

```python
df = df[["QUESTION", "CONTEXT", "ANSWER"]]
df.head()
```

```python
df.to_csv("process_questions_601_700.csv", index=False)
```
